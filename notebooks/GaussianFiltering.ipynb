# Calculate first degree derivative

A sharp change in function $f(x)$ corresponds to a peak in its derivate $f'(x)$ which in turn corresponds to 
a zero crossing in $f''(x)$.

Using Taylor expansion:

\begin{align*}
f(x+h) &= f(x) + hf'(x) + \frac{1}{2}h^2f''(x) + O(h^3)\\
f(x-h) &= f(x) - hf'(x) + \frac{1}{2}h^2f''(x) + O(h^3)\\
f'(x) &= \frac{-f(x-h)+f(x+h)}{2h}
f''(x) &= \frac{f(x-h)-2f(x)+f(x+h)}{2h^2}\\
\end{align*}


So a first degree derivative corrsponds to convolution with $\begin{pmatrix}-1 & 0 & 1  \end{pmatrix}$
while a second degree derivative will correspond to convolution with $\begin{pmatrix}1 & -2 & 1  \end{pmatrix}$
